```python
#import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
df = df.iloc[:,1:]
```

```python
# scatter plot petal length  vs petal width
df.plot.scatter(x='PetalLengthCm', y='PetalWidthCm')
plt.show()
```

```python
#scatter plot sepal length vs sepal width
df.plot.scatter(x='SepalLengthCm',y='SepalWidthCm')
plt.show()
```

```python
sns.set_style('whitegrid');
sns.scatterplot(data=df,x='PetalLengthCm',y='PetalWidthCm',hue='Species')
plt.legend()
plt.show()
```

```python
# pair plot
sns.set_style('whitegrid')
sns.pairplot(df,hue='Species',size=3)
plt.show()
```

```python
# pdf of petal length
sns.set_style('whitegrid')
sns.displot(df,x='PetalLengthCm',hue='Species',kind='kde',fill='True')
plt.legend()
plt.show()
```

```python
#cdf of petal length
sns.ecdfplot(data=df, x="PetalLengthCm",hue='Species')
plt.show()
```

```python
df.describe()
```

```python
#box plot petal length
sns.boxplot(data=df,x='Species', y='PetalLengthCm')
plt.show()
```

```python
# violin plot petal length
sns.violinplot(data=df, x='Species', y='PetalLengthCm', size=8)
plt.show()
```

```python
sns.jointplot(data=df, x='PetalLengthCm',y='PetalWidthCm',hue='Species')
plt.show()
```

```python
sns.jointplot(data=df, x='PetalLengthCm',y='PetalWidthCm',kind='kde',hue='Species')
plt.show()
```
